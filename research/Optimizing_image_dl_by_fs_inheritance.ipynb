# Optimising image in-depth analysis by grouping per images base image
This notebook aims at grouping the docker images per base fs layer in order to optimize their download speed across nodes. See [docker fs layer documentation](https://docs.docker.com/storage/storagedriver/#images-and-layers). It was used for the [docker analysis script](https://github.com/src-d/docker-image-analysis).
## Data Loading

```python
import json
import pandas as pd
import matplotlib
%matplotlib inline

with open('images_final.txt', 'r') as f:
    images = f.read().split("\n")

len(images)
```

```python
def build_path(image):
    folder = image[:2]
    return f"./manifests/{folder}/{image.replace(':latest', '')}.json"

build_path(images[0]) 
```

```python
def process_layers(manifest_data):
    layers = manifest_data["layers"]
    hashes = [layer["digest"] for layer in layers]
    return hashes

images_fs = {}
other_images = []
for image in images:
    try:
        with open(build_path(image)) as manifest:  
            manifest_data = json.load(manifest)
            layers =  process_layers(manifest_data)
            assert len(layers)
            images_fs[image] = layers
            
    except Exception as e:
        other_images.append(image)
len(images_fs)
```

```python
images_fs_sliced = []
for image in images_fs:
    images_fs_sliced.append((image, images_fs[image][0]))

images_fs_sliced.sort(key=lambda img:img[1])
prev = ""
files = [open(f"./image{count%4}","w+") for count in range(4)]
count = 0
print(len(images_fs_sliced))
for image in images_fs_sliced:
    if image[1] != prev:
        count += 1
        prev = image[1]
    files[count%4].write(f"{image[0]}\n")
print(count)
```
